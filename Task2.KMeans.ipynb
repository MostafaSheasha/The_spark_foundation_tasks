## Importing libraries & data

```python
import pandas as pd  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
df = pd.read_csv("Iris.csv")
df=df.drop(['Id','Species'],axis=1)
df.head()
```

```python
df.shape
```

## The elbow method

```python
from sklearn.cluster import KMeans
css = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    css.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.figure(figsize=(10,6))
plt.style.use('bmh')
plt.plot(range(1, 11), css,'o--')
plt.title('The elbow method',fontsize=18)
plt.xlabel('Number of clusters',fontsize=15)
plt.ylabel('Within cluster sum of squares',fontsize=15)
plt.show();
```

From this plot we find the optimum number of clusters is '3'.

## Kmeans model with 3 clusters

```python
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(df)
```

```python
x = df.values
plt.figure(figsize=(12,8))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'virginica')

plt.legend()
```
